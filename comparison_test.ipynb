```python
from tinysmpc import VirtualMachine, PrivateScalar, SharedScalar
from tinysmpc.fixed_point import fixed_point, float_point
import random
```

```python
client = VirtualMachine('client')
clique = VirtualMachine('clique')
```

```python
#Generate 1000 random numbers between 1 and 30
data = []
for i in range(0,1000):
    n = random.randint(1,30)
    data.append(n)
#convert data to private data of client1
private_data = []
for n in data:
    private_data.append(PrivateScalar(n,client))
```

```python
print(client)
print(clique) #Clique can not see the private data of client1
```

```python
shared_data = []
#client1 performs secret sharing
for pd in private_data:
    sd = pd.share([client,clique])
    shared_data.append(sd)
for i in range(5):
    print(shared_data[i])
```

```python
#secure comparison with a public integer
threshold = 15
results = []
for sd in shared_data:
    res = (sd > threshold)
    results.append(res)

for i in range(5):
    print(results[i])
```

```python
#secure comparison between float numbers
float_data = []
for i in range(100):
    x = random.random()
    x = round(x, 8)
    float_data.append(x)
```

```python
shared_float_data = []
for d in float_data:
    shared_f = PrivateScalar(fixed_point(d),client).share([client,clique])
    shared_float_data.append(shared_f)
```

```python
shared_float_data
```

```python
thres_float = int(0.5 * 1e8) #multiply the threshold 0.5 by 10^8
results_float = []
for d in shared_float_data:
    res = (d > thres_float)
    results_float.append(res)

for i in range(5):
    print(results_float[i])
```
